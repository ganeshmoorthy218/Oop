```python
##E-Commerce Analytics Project##
##Install + Imports
import sqlite3
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

fake = Faker()
```

```python
##Create SQLite DB + Schema
conn = sqlite3.connect('ecommerce.db')
cursor = conn.cursor()

cursor.executescript("""
DROP TABLE IF EXISTS payments;
DROP TABLE IF EXISTS order_items;
DROP TABLE IF EXISTS orders;
DROP TABLE IF EXISTS products;
DROP TABLE IF EXISTS categories;
DROP TABLE IF EXISTS customers;

CREATE TABLE customers (
    customer_id INTEGER PRIMARY KEY,
    name TEXT,
    email TEXT UNIQUE,
    city TEXT,
    country TEXT,
    signup_date DATE
);

CREATE TABLE categories (
    category_id INTEGER PRIMARY KEY,
    category_name TEXT
);

CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    product_name TEXT,
    price REAL,
    category_id INTEGER,
    FOREIGN KEY(category_id) REFERENCES categories(category_id)
);

CREATE TABLE orders (
    order_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    order_date DATE,
    status TEXT,
    FOREIGN KEY(customer_id) REFERENCES customers(customer_id)
);

CREATE TABLE order_items (
    order_item_id INTEGER PRIMARY KEY,
    order_id INTEGER,
    product_id INTEGER,
    quantity INTEGER,
    FOREIGN KEY(order_id) REFERENCES orders(order_id),
    FOREIGN KEY(product_id) REFERENCES products(product_id)
);

CREATE TABLE payments (
    payment_id INTEGER PRIMARY KEY,
    order_id INTEGER,
    payment_method TEXT,
    payment_amount REAL,
    FOREIGN KEY(order_id) REFERENCES orders(order_id)
);
""")

conn.commit()
print("Schema created")
```

```python
##Insert Categories & Products
categories = ["Electronics","Laptops","Mobiles","Accessories"]
for cat in categories:
    cursor.execute("INSERT INTO categories (category_name) VALUES (?)", (cat,))

products = []
for i in range(1, 41):
    category_id = random.randint(1, len(categories))
    price = round(random.uniform(200, 2000), 2)
    products.append((fake.word().capitalize()+" "+fake.word().capitalize(), price, category_id))

cursor.executemany("""
INSERT INTO products (product_name, price, category_id)
VALUES (?, ?, ?)
""", products)

conn.commit()
print("Product & category data inserted")
```

```python
##Insert Customers
customers = []
# Use a set to keep track of generated emails to ensure uniqueness
generated_emails = set()
for _ in range(500):
    while True:
        email = fake.email()
        if email not in generated_emails:
            generated_emails.add(email)
            break
    customers.append((fake.name(), email,
                      fake.city(), fake.country(),
                      fake.date_between(start_date="-2y", end_date="today")))

cursor.executemany("""
INSERT INTO customers (name, email, city, country, signup_date)
VALUES (?, ?, ?, ?, ?)
""", customers)
conn.commit()
print("Customers inserted ")
```

```python
##Generate Orders + Items + Payments
statuses = ["Completed","Pending","Cancelled"]
payment_methods = ["Credit Card","UPI","Wallet","Cash on Delivery"]

orders = []
order_items = []
payments = []

order_id = 1
item_id = 1
payment_id = 1

for _ in range(5000):
    cust_id = random.randint(1, 500)
    order_date = fake.date_between(start_date="-1y", end_date="today")
    status = random.choice(statuses)

    orders.append((order_id, cust_id, order_date, status))

    for _ in range(random.randint(1,4)):
        product_id = random.randint(1, 40)
        qty = random.randint(1, 3)
        order_items.append((item_id, order_id, product_id, qty))

        price = cursor.execute("SELECT price FROM products WHERE product_id = ?", (product_id,)).fetchone()[0]
        payments.append((payment_id, order_id, random.choice(payment_methods), price*qty))

        item_id += 1
        payment_id += 1

    order_id += 1

cursor.executemany("INSERT INTO orders VALUES (?, ?, ?, ?)", orders)
cursor.executemany("INSERT INTO order_items VALUES (?, ?, ?, ?)", order_items)
cursor.executemany("INSERT INTO payments VALUES (?, ?, ?, ?)", payments)

conn.commit()
print("Orders, order items, and payments inserted ")
```

```python
##Monthly Revenue Trend
query = """
SELECT strftime('%Y-%m', order_date) AS month,
       SUM(payment_amount) AS total_revenue
FROM orders
JOIN payments USING(order_id)
GROUP BY month
ORDER BY month
"""
df = pd.read_sql_query(query, conn)
df
df.plot(x='month', y='total_revenue', kind='line', figsize=(10,5), title='Revenue Trend')
plt.show()
```

```python
##Top 10 Best-Selling Products
query = """
SELECT p.product_name,
       SUM(oi.quantity) AS total_qty
FROM products p
JOIN order_items oi USING(product_id)
GROUP BY p.product_id
ORDER BY total_qty DESC
LIMIT 10
"""
pd.read_sql_query(query, conn)
```

```python
##Top 10 High-Value Customers (LTV)
query = """
SELECT c.name,
       c.city,
       SUM(pay.payment_amount) AS lifetime_value
FROM customers c
JOIN orders o USING(customer_id)
JOIN payments pay USING(order_id)
GROUP BY c.customer_id
ORDER BY lifetime_value DESC
LIMIT 10
"""
pd.read_sql_query(query, conn)
```

```python
##Category-wise Revenue Share
query = """
SELECT cat.category_name,
       SUM(pay.payment_amount) AS revenue
FROM categories cat
JOIN products p USING(category_id)
JOIN order_items oi USING(product_id)
JOIN payments pay USING(order_id)
GROUP BY cat.category_id
"""
df2 = pd.read_sql_query(query, conn)
df2
df2.plot(x="category_name", y="revenue", kind="bar", figsize=(8,5))
plt.show()
```
